```python
import socket
import threading
import tkinter as tk
from tkinter import scrolledtext
from tkinter import messagebox


def receive_messages(sock, text_area):
    while True:
        try:
            message = sock.recv(1024).decode('utf-8')
            if not message:
                break
            text_area.config(state=tk.NORMAL)
            text_area.insert(tk.END, "Received: " + message + "\n")
            text_area.config(state=tk.DISABLED)
            text_area.see(tk.END)  # Auto-scroll to the bottom
        except OSError:  # Possibly client disconnected
            break
    sock.close()
    print("Connection closed.")


def send_message(sock, entry, text_area):
    message = entry.get()
    if message:
        try:
            sock.sendall(message.encode('utf-8'))
            text_area.config(state=tk.NORMAL)
            text_area.insert(tk.END, "Sent: " + message + "\n")
            text_area.config(state=tk.DISABLED)
            text_area.see(tk.END)  # Auto-scroll to the bottom
            entry.delete(0, tk.END)
        except OSError:
            messagebox.showerror("Error", "Connection lost.")


def connect_to_server(host, port, text_area):
    try:
        sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        sock.connect((host, port))

        text_area.config(state=tk.NORMAL)
        text_area.insert(tk.END, "Connected to server\n")
        text_area.config(state=tk.DISABLED)

        receive_thread = threading.Thread(target=receive_messages, args=(sock, text_area))
        receive_thread.daemon = True  # Allow main thread to exit
        receive_thread.start()
        return sock
    except ConnectionRefusedError:
        messagebox.showerror("Error", "Connection refused. Check server status and port.")
        return None
    except OSError as e:
        messagebox.showerror("Error", f"Connection error: {e}")
        return None


def on_closing():
    if sock:
        sock.close()  # Close socket before closing the window
    window.destroy()


window = tk.Tk()
window.title("Socket Client")
window.protocol("WM_DELETE_WINDOW", on_closing) #Proper exit handling

host_label = tk.Label(window, text="Server IP:")
host_label.grid(row=0, column=0, padx=5, pady=5)
host_entry = tk.Entry(window)
host_entry.grid(row=0, column=1, padx=5, pady=5)
host_entry.insert(0, "127.0.0.1")  # Default localhost

port_label = tk.Label(window, text="Port:")
port_label.grid(row=1, column=0, padx=5, pady=5)
port_entry = tk.Entry(window)
port_entry.grid(row=1, column=1, padx=5, pady=5)
port_entry.insert(0, "12345")  # Default port


text_area = scrolledtext.ScrolledText(window, wrap=tk.WORD)
text_area.grid(row=2, column=0, columnspan=2, padx=5, pady=5)
text_area.config(state=tk.DISABLED) #Initially make it read-only

message_entry = tk.Entry(window)
message_entry.grid(row=3, column=0, padx=5, pady=5)


send_button = tk.Button(window, text="Send", command=lambda: send_message(sock, message_entry, text_area))
send_button.grid(row=3, column=1, padx=5, pady=5)

connect_button = tk.Button(window, text="Connect", command=lambda: connect_to_server(host_entry.get(), int(port_entry.get()), text_area))
connect_button.grid(row=4, column=0, columnspan=2, padx=5, pady=5)


sock = None # Global variable for the socket
window.mainloop()
```
